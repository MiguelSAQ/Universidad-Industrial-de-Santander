```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
S1 = (1.1, -1.6)
S2 = (1.0, -2.8)
S3 = (1.1, -3.4)
S4 = (1.7, -3.6)
S5 = (1.6, -4.3)
S6 = (1.3, -4.5)
S7 = (1.3, -6.4)
S8 = (1.0, 5.5)
S9 = (1.1, 4.5)
S10 = (1.1, 1.5)

Lanzamientos = np.array([S1, S2, S3, S4, S5, S6, S7, S8, S9, S10])
Prom_lanzamientos = np.mean(Lanzamientos, axis=0)
Std_lanzamientos = np.std(Lanzamientos, axis=0)
                         
# Códigos de formato ANSI
negrita = "\033[1m"
fin_formato = "\033[0m"

# Títulos en negrita
print(negrita + "Impacto promedio:" + fin_formato)
print(Prom_lanzamientos, "cm")
print()
print(negrita + "Desviacion estandar del impacto promedio" + fin_formato)
print(Std_lanzamientos, "cm")
```

```python
plt.figure(figsize=(8,8))

r = np.linspace(1.25, 6.25, 5)

for i in range(5):
    # Crear un conjunto de puntos en el círculo
    theta = np.linspace(0, 2*np.pi, 100)
    x = r[i] * np.cos(theta)
    y = r[i] * np.sin(theta)

    # Trazar el círculo con color negro claro
    plt.plot(x, y, color='lightgray')

x_5 = 5 * 1.25 * np.cos(theta)
y_5 = 5 * 1.25 * np.sin(theta)
plt.fill_between(x_5, y_5, color='lightgray', alpha=0.5)
x_4 = 4 * 1.25 * np.cos(theta)
y_4 = 4 * 1.25 * np.sin(theta)
plt.fill_between(x_4, y_4, color='white', alpha=0.5)
x_3 = 3 * 1.25 * np.cos(theta)
y_3 = 3 * 1.25 * np.sin(theta)
plt.fill_between(x_3, y_3, color='lightgray', alpha=0.5)
x_2 = 2 * 1.25 * np.cos(theta)
y_2 = 2 * 1.25 * np.sin(theta)
plt.fill_between(x_2, y_2, color='white', alpha=0.5)
x_1 = 1 * 1.25 * np.cos(theta)
y_1 = 1 * 1.25 * np.sin(theta)
plt.fill_between(x_1, y_1, color='lightgray', alpha=0.5)

# Graficar los puntos en la diana
plt.scatter(*zip(S1, S2, S3, S4, S5, S6, S7, S8, S9, S10), marker='x', color='black', s=40, label='Impactos')

plt.title('Impactos en la diana', fontweight='bold')
plt.xlabel('Eje X (cm)')
plt.ylabel('Eje Y (cm)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.axis('equal')  
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def cuadratica(x):
    return -0.19 * x**2 + 0.36 * x + 0.18

x_values = np.linspace(0, 2.4, 100)
y_values = cuadratica(x_values)

# Coordenadas del punto de impacto
impacto_x = 0.95
impacto_y = 0.3349

# Margen de error en y
margen_error_y = 0.0380222303

# Ajustar el tamaño de la figura
plt.figure(figsize=(12, 3))  # Ancho x alto en pulgadas

plt.plot(x_values, y_values, label='Lanzamiento teórico', color='k', linestyle='--', linewidth=2)
plt.scatter(2.3,0, marker='o', color='Orange', s=100, label='Alcance Maximo = 2.30 m')
plt.errorbar(impacto_x, impacto_y, yerr=margen_error_y, fmt='o', color='blue', capsize=5, label='Impacto Promedio')
plt.scatter(0.95,0.35, marker='x', color='red', s=100, label='Impacto Esperado')

plt.title('Lanzamiento Teorico e Impacto Promedio', fontweight='bold')
plt.xlabel('Eje X (m)')
plt.ylabel('Eje Y (m)')
plt.legend()
plt.grid(True)
plt.xlim(0, 2.4) 
plt.ylim(0, 0.6)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
