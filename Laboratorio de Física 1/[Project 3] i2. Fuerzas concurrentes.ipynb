```python
import numpy as np
import matplotlib.pyplot as plt
import math as math
```

# 3 Masas

## Caso 1

```python
V11 = (-1.39, -1.39)
V21 = (1.39, -1.39)
V31 = (0, 2.83)
V31_min = (-0.2*1.11, 2.63*1.11)
V31_max = (0.2, 2.93)
V31_teo = [x + y for x, y in zip(V11, V21)]

fig, ax = plt.subplots()

ax.quiver(0, 0, V11[0], V11[1], angles='xy', scale_units='xy', scale=1, color='r', zorder=2, label=('F1 = 1.96 N'))
ax.quiver(0, 0, V21[0], V21[1], angles='xy', scale_units='xy', scale=1, color='r', zorder=2, label=('F2 = 1.97 N'))
ax.quiver(0, 0, V31_teo[0], -V31_teo[1], angles='xy', scale_units='xy', scale=1, color='g', zorder=2, label=('Ft = 2.78 N'))
ax.quiver(0, 0, V31[0], V31[1], angles='xy', scale_units='xy', scale=1, color='k', zorder=2, label=('Fr = 2.83 N'))

polygon_points1 = np.array([V31_min, V31_max, (0, 0)])
polygon1 = plt.Polygon(polygon_points1, closed=True, edgecolor='k', linestyle='--', linewidth=1.5, facecolor='gray', alpha=0.3, label='Error')
ax.add_patch(polygon1)

ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')

ax.set_xlim(-3, 3)
ax.set_ylim(-2, 4)

ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)

plt.grid(zorder=0)
plt.title(r'$\bf{3\ masas\ [Caso\ 1]}$')
plt.legend()
plt.show()
```

```python
VP31 = (math.sqrt(V31[0]**2 + V31[1]**2), math.degrees(math.atan2(V31[0], V31[1])))
VP31_Teo = (math.sqrt(V31_teo[0]**2 + V31_teo[1]**2), math.degrees(math.atan2(V31_teo[0], -V31_teo[1])))
VP31_min = (math.sqrt(V31_min[0]**2 + V31_min[1]**2), math.degrees(math.atan2(V31_min[0], V31_min[1])))
VP31_max = (math.sqrt(V31_max[0]**2 + V31_max[1]**2), math.degrees(math.atan2(V31_max[0], V31_max[1])))
Caso1 = np.array([VP31, VP31_min, VP31_min])
Caso1_std = np.std(Caso1, axis=0)

print('\033[1mMasa experimental:\033[0m')
print(round(VP31[0]*1000/9.8,2), 'g')
print()
print('\033[1mMasa teorica:\033[0m')
print(round(VP31_Teo[0]*1000/9.8,2), 'g')
print()
print('\033[1mError de la masa:\033[0m')
print(("± {:.2f}".format(round(Caso1_std[0]*1000/9.8 ,2))), 'g')
print()
print('\033[1mAngulo experimental:\033[0m')
print("{:.2f}°".format(round(180-VP31[1],0)))
print()
print('\033[1mAngulo teorico:\033[0m')
print("{:.2f}°".format(round(180-VP31_Teo[1],0)))
print()
print('\033[1mError del angulo:\033[0m')
print("± {:.2f}°".format(round(Caso1_std[1], 0)))
```
