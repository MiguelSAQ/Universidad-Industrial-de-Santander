# Densidad del Alumisnio

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

## Cubo

```python
# Datos del cubo
Mcubo = np.array([44.00, 44.03, 44.10])  # Masa
Vcubo = np.array([16.35, 16.55, 16.29])  # Volumen

# Cálculos
Mcubo_mean, Mcubo_std = np.mean(Mcubo), np.std(Mcubo)
Vcubo_mean, Vcubo_std = np.mean(Vcubo), np.std(Vcubo)
D_cubo = Mcubo_mean / Vcubo_mean
D_cubos = Mcubo / Vcubo
D_cubo_mean, D_cubo_std = np.mean(D_cubos), np.std(D_cubos)

# Resultados
resultados_cubo = {
    "Promedio de las masas del cubo": f"{Mcubo_mean:.2f} g",
    "Desviación estándar de las masas del cubo": f"{Mcubo_std:.2f} g",
    "Promedio de los volúmenes del cubo": f"{Vcubo_mean:.2f} cm³",
    "Desviación estándar de los volúmenes del cubo": f"{Vcubo_std:.2f} cm³",
    "Densidad del promedio de las masas entre el promedio de los volúmenes": f"{D_cubo_mean:.2f} g/cm³",
    "Densidades de cada dato por separado": [f"{densidad:.2f} g/cm³" for densidad in D_cubos],
    "Promedio de las densidades": f"{D_cubo_mean:.2f} g/cm³",
    "Desviación estándar de las densidades": f"{D_cubo_std:.2f} g/cm³",
}

# Imprimir resultados
for titulo, valor in resultados_cubo.items():
    print(f"\033[1m{titulo}:\033[0m\n{valor}\n")
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(Vcubo, Mcubo, color='blue', label='Datos del Cubo')
plt.title('Gráfico de Masa vs Volumen [Cubo]')
plt.xlabel('Volumen (cm³)')
plt.ylabel('Masa (g)')
plt.legend()
plt.grid(True)
plt.ylim(43,45.10)
plt.show()
```
