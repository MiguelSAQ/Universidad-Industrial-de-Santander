```python
# Energia Final Total - Tabla 1

Ef1 = []
Ef1error = []
    
for i in range(len(D_H1)):
    Efinal1 = 0.5 * m_cte * (Vf1[i]**2)
    Efinal1error = 0.5 * m_cte * (Vf1std[i]**2)
    Ef1.append(Efinal1)
    Ef1error.append(Efinal1error)
    
print("\033[31m\033[1mEnergia Final Total - Tabla 1\033[0m")
print(Ef1)
print(Ef1error)
print()
```

```python
# Cambio en la Energia - Tabla 1

DE1 = []
    
for i in range(len(D_H1)):
    CambioEnergia1 = Ef1[i] - Ei1[i]
    DE1.append(CambioEnergia1)
    

print("\033[31m\033[1mCambio en la Energia Total - Tabla 1\033[0m")
print(DE1)
print()
```

```python
# Energia Potencial incial - Tabla 1

E_ui1 = []

for i in range(len(D_H1)):
    Epotencial1 = m_cte * g * D_H1[i]
    E_ui1.append(Epotencial1)
    
print("\033[31m\033[1mEnergia Potencial incial - Tabla 1\033[0m")
print(E_ui1)
print()
```

```python
# Energia Cinetica inicial - Tabla 1

E_ki1 = []
    
for i in range(len(D_H1)):
    Ecineticainicial1 = 0.5 * m_cte * (Vi1[i]**2)
    E_ki1.append(Ecineticainicial1)
    
print("\033[31m\033[1mEnergia Cinetica inicial - Tabla 1\033[0m")
print(E_ki1)
print()
```

```python
# Energia Cinetica final - Tabla 1

E_kf1 = []
    
for i in range(len(D_H1)):
    Ecineticafinal1 = 0.5 * m_cte * (Vf1[i]**2)
    E_kf1.append(Ecineticafinal1)
    
print("\033[31m\033[1mEnergia Cinetica final - Tabla 1\033[0m")
print(E_kf1)
print()
```
