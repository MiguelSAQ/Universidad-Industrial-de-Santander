```python
a = []
for j in range(1, 10):
    v1 = globals()[f'v1{j}']
    v2 = globals()[f'v2{j}']
    t = globals()[f't{j}']
    a.append([(v2[i] - v1[i]) / t[i] for i in range(3)])
```

```python
for i in range(1, 10):
    globals()[f'a{i}'] = a[i-1]
    globals()[f'std_a{i}'] = np.std(a[i-1])
    globals()[f'mean_a{i}'] = np.mean(a[i-1])
    mean_formatted = "\033[1m{:.4f} m/s²\033[0m".format(globals()[f'mean_a{i}'])
    std_formatted = "\033[1m± {:.4f} m/s²\033[0m".format(globals()[f'std_a{i}'])
    print(f'La aceleración promedio fue de: {mean_formatted} con una desviación estándar de: {std_formatted}')
```
