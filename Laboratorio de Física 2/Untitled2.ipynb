```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
F = np.array([0.8, 1.125, 1.50, 1.6, 1.75]) * 1e15  # Multiplicando por 10^15
E = np.array([0, 1, 2, 3, 4]) * 1.60218e-19  # Multiplicando por 1.60218*10^{-19}

# Ajuste lineal usando polyfit (grado 1 para una recta)
coef = np.polyfit(F, E, 1)  # Ajustar E en función de F
slope, intercept = coef

# Crear la recta de ajuste
E_ajustado = np.polyval(coef, F)

# Graficar los datos y la recta de ajuste
plt.scatter(F, E, label="Datos experimentales", color="blue")
plt.plot(F, E_ajustado, label=f"Ajuste lineal: m = 6.36 x 10⁻³⁴", color="red")

plt.xlabel('F (Hz)')
plt.ylabel('E (J)')
plt.title('Ajuste lineal de E vs F')
plt.legend()
plt.grid(True)
plt.show()

# Mostrar los parámetros de la recta
print(f'Pendiente (experimental):  6.36 x 10⁻³⁴ J⋅s')
print(f'Pendiente (teorica):   6.62 x 10⁻³⁴ J⋅s')

# Valor teórico de la constante de Planck
h_teorico = 6.626e-34  # J·s

# Calcular el error entre la pendiente experimental y la teórica
error = abs((slope - h_teorico) / h_teorico) * 100
print(f'Error entre la pendiente experimental y la teórica: {error:.2f}%')
```
