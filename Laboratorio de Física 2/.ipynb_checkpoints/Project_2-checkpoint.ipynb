# Proyecto 2. Materiales óhmicos y no-óhmicos

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Medición 1 

```python
# Resistencia constante: 27.6
V1 = np.array([39.50, 35.85, 30.26, 24.60, 18.40, 12.40, 7.82])  # Voltaje (V)
C1 = np.array([1.42, 1.26, 1.05, 0.86, 0.63, 0.43, 0.26])  # Corriente (A)
```

```python
# Ajuste lineal
coef1 = np.polyfit(C1, V1, 1)
fit_line1 = np.polyval(coef1, C1)
x1 = np.linspace(0.251, 1.41, 100)
y1 = 27.6 * x1
err1 = (np.abs(coef1[0]-27.6)/27.6)*100

# Grafica
plt.figure(figsize=(8, 6))
plt.scatter(C1, V1, marker='o', color='k', label='Datos experimentales')
plt.plot(C1, fit_line1, linestyle='--', color='blue', label=f'Ajuste lineal      (m={coef1[0]:.2f})')
plt.plot(x1, y1, linestyle='--', color='red', label='Gráfica teórica  (m=27.6)')

plt.title("Gráfica de el Voltaje vs  la Corriente con la Resistencia Constante", fontweight='bold')

plt.xlabel("Corriente (A)")
plt.ylabel("Voltaje (V)")

plt.grid()
plt.legend()
plt.show()

print("• Ecuacion teorica: y = (27.60)x ")
print(f"• Ecuacion experimental: y = ({coef1[0]:.4g})x")
print(f"• El error obtenido en la medición 1 fue de {err1:.2g} %")
```

## Medición 2

```python
# Voltaje constante: 13.52
R2 = np.array([10, 20, 30, 40, 50, 60, 70])
C2 = np.array([1.32, 0.69, 0.47, 0.34, 0.27, 0.22, 0.19])
inv_R2 = 1 / R2
```

```python
# Ajuste
coef2 = np.polyfit(inv_R2, C2,1)
pol2 = np.poly1d(coef2)
inv_R2_fit = np.linspace(inv_R2.min(), inv_R2.max(), 100)
V_fit2 = pol2(inv_R2_fit)
R_fit2 = 1 / inv_R2_fit

# Grafica
plt.figure(figsize=(8, 6))
plt.scatter(R2, C2, marker='o', color='k', label='Datos experimentales')
plt.plot(R_fit2, V_fit2, linestyle='--', color='blue', label='Ajuste [1/R]')

plt.title("Gráfica de la Corriente vs la Resistencia con el Voltaje Constante", fontweight='bold')

plt.xlabel('Resistencia (Ω)')
plt.ylabel('Corriente (A)')

plt.legend()
plt.grid(True)
plt.show()
```

```python
#Ajuste lineal
coef2 = np.polyfit(inv_R2, C2, 1)
fit_line2 = np.polyval(coef2, inv_R2)
x2 = np.linspace(0.015, 0.1, 100)
y2 = 13.52 * x2
err2 = (np.abs(coef2[0]-13.52)/13.52)*100

# Grafica
plt.figure(figsize=(8, 6))
plt.scatter(inv_R2, C2, marker='o', color='k', label='Datos experimentales')
plt.plot(inv_R2, fit_line2, linestyle='--', color='blue', label=f'Ajuste lineal      (m={coef2[0]:.2f})')
plt.plot(x2, y2, linestyle='--', color='red', label=f'Gráfica teórica  (m=13.52')

plt.title("Gráfica de la Corriente vs el Inverso de la Resistencia", fontweight='bold')

plt.xlabel("Inverso de la Resistencia (1/Ω)")
plt.ylabel('Corriente (A)')

plt.legend()
plt.grid(True)
plt.show()

print("• Ecuacion teorica: y = (13.52)x ")
print(f"• Ecuacion experimental: y = ({coef2[0]:.4g})x")
print(f"• El error obtenido en la medición 2 fue de {err2:.2g} %")
```

## Medición 3

```python
V2 = np.array([10.82, 15.20, 20.60, 26.90, 40.00, 67.20, 79.20, 99.90])
C2 = np.array([00.16, 00.18, 00.22, 00.25, 00.32, 00.42, 00.47, 00.53])
R2 = np.array([19.20, 51.00, 119.20, 144.10, 177.00, 202.5, 194.8, 195.7])
```

```python
coef21 = np.polyfit(C2, V2, 2)
fit_curve21 = np.poly1d(coef21)

C2_linea21 = np.linspace(min(C2), max(C2), 100)
V2_linea21 = fit_curve21(C2_linea21)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(C2, V2, marker='o', color='k', label='Datos experimentales')
plt.plot(C2_linea21, V2_linea21, linestyle='--', color='blue', label='Ajuste cuadrático')

plt.title("Gráfica de el Voltaje vs la Corriente de un Bombillo", fontweight='bold')

plt.xlabel("Corriente (A)")
plt.ylabel("Voltaje (V)")

plt.grid()
plt.legend()
plt.show()
```

```python
coef22 = np.polyfit(C2, R2, 4)
fit_curve22 = np.poly1d(coef22)

C2_linea22 = np.linspace(min(C2), max(C2), 100)
V2_linea22 = fit_curve22(C2_linea22)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(C2, R2, marker='o', color='k', label='Datos experimentales')
plt.plot(C2_linea22, V2_linea22, linestyle='--', color='blue', label='Ajuste')

plt.title("Gráfica de la Corriente vs la Resistencia de un Bombillo", fontweight='bold')

plt.xlabel("Corriente (A)")
plt.ylabel("Resistencia (Ω)")

plt.grid()
plt.legend()
plt.show()
```
