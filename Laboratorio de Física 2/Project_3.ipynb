## Parte 1 (Constantán, 0.4mm)

### Fase 1: Medición directa de la resistencia

```python
# Datos
L_A_P11 =  np.array([795774.881,1591549.762,2387324.643,3183099.524,3978874.405,4774649.286,5570424.167,6366199.048,7161973.929,7957748.81])
R_P11 = np.array([1.2,1.5,1.9,2.4,2.7,3,3.6,4,4.3,4.4])

# Ajuste
coef_P11 = np.polyfit(L_A_P11, R_P11, 1)
fit_P11 = np.polyval(coef_P11, L_A_P11)
X_P11 = np.linspace(0, 7957748.81, 100)
Y_Exp_P11 = X_P11 * coef_P11[0] + coef_P11[1]
Y_Teo_P11 = X_P11 * 4.9*10**(-7) + coef_P11[1]

# Grafica
plt.figure(figsize=(7, 5))
plt.scatter(L_A_P11, R_P11, marker='o', color='k', label='Datos experimentales')
plt.plot(X_P11, Y_Exp_P11, linestyle='--', color='r', label=f'Resistividad experimental = 4.78 × 10⁻⁷')
plt.plot(X_P11, Y_Teo_P11, linestyle='--', color='b', label=f'Resistividad teorica = 4.90 × 10⁻⁷ ')
plt.title("Medición Directa de la Resistencia (Constantán, 0.4mm)", fontweight='bold')
plt.xlabel("L/A  (1/m)")
plt.ylabel("Resistencia (Ω)")
plt.grid()
plt.legend()
plt.text(0.95, 0.05, f'Error:  {np.mean(Err_P11):.2f} %', transform=plt.gca().transAxes, 
         fontsize=12, verticalalignment='bottom', horizontalalignment='right',
         bbox=dict(facecolor='white', alpha=0.7, edgecolor='black'))
plt.show()

# Error
Err_P11 = np.zeros(len(R_P11))
for i in range(len(R_P11)):
    Err_P11[i] = np.abs( 4.9*10**(-7) * L_A_P11[i] + coef_P11[1] - R_P11[i] ) / ( 4.9*10**(-7) * L_A_P11[i] ) * 100
print(f"El error promedio fue de: {np.mean(Err_P11):.2f} %")
```
