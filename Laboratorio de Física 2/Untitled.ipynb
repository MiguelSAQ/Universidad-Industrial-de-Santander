```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
PRTCM = np.array([0.0000001, 0.8, 1.6, 3.2, 4.8, 6.5, 8.1, 9.7, 11.3, 12.9, 14.5, 16.1, 17.7, 19.3]) #x
RI = np.array([1.00, 0.95, 0.80, 0.39, 0.079, 0.003, 0.036, 0.043, 0.013, 0.0003, 0.012, 0.015, 0.0044, 0.0003]) #y

for i in range(0, 14, 1):
    Intensidad[i] = ((PRTCM[i])/(np.sin(PRTCM[i]/2)))**(-2)
    
print(Intensidad[0])

I_max = np.polyfit(Intensidad, RI, 1)
print(I_max[0])
```

```python
PRTCM = np.array([0.00000001, 0.8, 1.6, 3.2, 4.8, 6.5, 8.1, 9.7, 11.3, 12.9, 14.5, 16.1, 17.7, 19.3])
RI = np.array([1.00, 0.95, 0.80, 0.39, 0.079, 0.003, 0.036, 0.043, 0.013, 0.0003, 0.012, 0.015, 0.0044, 0.0003])

PRTCM_teo = np.linspace(0.00000001, 19.3, 100)
RI_teo = np.zeros(100)

for i in range(100):
    RI_teo[i] = I_max[0] * (((np.sin(PRTCM_teo[i]/2))**2)/((PRTCM_teo[i]))**2)
    
plt.figure(figsize=(8, 6))
plt.scatter(PRTCM, RI, label="Datos experimentales", color="red")
plt.plot(PRTCM_teo,RI_teo, label="Función de interpolación", color="Blue", linestyle = "--")
plt.title("Relative intensity as a function of distance from the central maximum", fontweight='bold')
plt.xlabel("Position Relative to Central Maximum (mm)")
plt.ylabel("Relative Intensity")
plt.grid()
plt.legend()
plt.show()
```

```python
PRTCM = np.array([0.00000001, 0.8, 1.6, 3.2, 4.8, 6.5, 8.1, 9.7, 11.3, 12.9, 14.5, 16.1, 17.7, 19.3])
RI = np.array([1.00, 0.95, 0.80, 0.39, 0.079, 0.003, 0.036, 0.043, 0.013, 0.0003, 0.012, 0.015, 0.0044, 0.0003])

ECM = 0

for i in range(14):
    RI_teo = I_max[0] * (np.sin(PRTCM[i] / 2)**2) / (PRTCM[i]**2)
    ECM += (1/14) * (RI[i] - RI_teo) ** 2
    
print(ECM)
```
