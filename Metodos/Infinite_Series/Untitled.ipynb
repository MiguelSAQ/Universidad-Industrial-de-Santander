### EXERCISE 5.1.2

Show that

$$ \sum_{n=1}^{\infty} \frac{1}{(2n-1)(2n+1)} = \frac{1}{2}$$

```python
while True:
    try:
        i = int(input("Ingrese un número entero positivo hasta donde quiere sumar: "))
        if i > 0:
            break
        else:
            print("Por favor, ingrese un entero mayor que 0.")
    except ValueError:
        print("Entrada inválida. Ingrese un número entero.")

R_1 = 0  # Resultado por sumatoria.
X = []   # Lista donde se almacenaran terminos "n". 
Y = []   # Lista donde se almacenaran el valor parcial de la sumatoria.

for n in range(1, i + 1):  # Sumatoria
    R_1 += 1 / ( ( 2*n-1 ) * (2*n+1) )  
    Y.append(R_1)
    X.append(n)

error_rel = abs(R_1 - 0.5) / 0.5 *100  

print("\n--- Resultados ---")
print(f"Resultado mediante la sumatoria: {R_1}")
print(f"Error relativo: {error_rel:.2e} %")


plt.plot(X,Y,linestyle="--",color="blue")
plt.axhline(y=0.5, color='red', linestyle='--', label='y = 1/2')
plt.title('Aproximación de la sumatoria hacia 1/2')
plt.xlabel('n (número de términos)')
plt.ylabel('Valor parcial de la sumatoria')
plt.ylim(0.33,0.55)
plt.grid(True)
plt.legend()
plt.show()
```

### EXERCISE 5.3.1

Show that

$$ \sum_{n=1}^{\infty} \frac{1}{n(n+1)} = 1$$

```python
while True:
    try:
        i = int(input("Ingrese un número entero positivo hasta donde quiere sumar: "))
        if i > 0:
            break
        else:
            print("Por favor, ingrese un entero mayor que 0.")
    except ValueError:
        print("Entrada inválida. Ingrese un número entero.")

R_1 = 0  # Resultado por sumatoria.
X = []   # Lista donde se almacenaran terminos "n". 
Y = []   # Lista donde se almacenaran el valor parcial de la sumatoria.

for n in range(1, i + 1):  # Sumatoria
    R_1 += 1 / ( n * (n+1) )  
    Y.append(R_1)
    X.append(n)

error_rel = abs(R_1 - 0.5) / 0.5 *100  

print("\n--- Resultados ---")
print(f"Resultado mediante la sumatoria: {R_1}")
print(f"Error relativo: {error_rel:.2e} %")


plt.plot(X,Y,linestyle="--",color="blue")
plt.axhline(y=1, color='red', linestyle='--', label='y = 1')
plt.title('Aproximación de la sumatoria hacia 1')
plt.xlabel('n (número de términos)')
plt.ylabel('Valor parcial de la sumatoria')
plt.ylim(0.5,1.2)
plt.grid(True)
plt.legend()
plt.show()
```
